```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

(a) Read Audio and Print Basic Information

```python
# Read the .wav file
sampling_rate, signal = wavfile.read("testing_voice.wav")  # replace with your file name

# If stereo, convert to mono
if len(signal.shape) > 1:
    signal = signal[:, 0]

# Number of samples
num_samples = len(signal)

# Duration (seconds)
duration = num_samples / sampling_rate

# Max and Min amplitude
max_amp = np.max(signal)
min_amp = np.min(signal)

# Print results
print("Sampling Rate:", sampling_rate, "Hz")
print("Number of Samples:", num_samples)
print("Duration:", duration, "seconds")
print("Maximum Amplitude:", max_amp)
print("Minimum Amplitude:", min_amp)
```

(b) Plot the Waveform

```python
# Time axis
time = np.linspace(0, duration, num_samples)

plt.figure()
plt.plot(time, signal)
plt.title("Waveform of Speech Signal")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

(c) Compute and Plot Magnitude Spectrum Using FFT

```python
# Compute FFT
fft_values = np.fft.fft(signal)

# Magnitude spectrum
magnitude = np.abs(fft_values)

# Frequency axis
freq = np.fft.fftfreq(num_samples, 1/sampling_rate)

# Plot only positive frequencies
half_n = num_samples // 2

plt.figure()
plt.plot(freq[:half_n], magnitude[:half_n])
plt.title("Magnitude Spectrum of Speech Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid()
plt.show()
```
